```python
import pandas as pd

Employee = pd.DataFrame({
    'Name': ['Atul','Raj','Darpan','Anmol','Piyush'],
    'Age': [25,30,23,22,32],
    'Designation': ['Manager','Clerk','Analyst','Clerk','Manager'],
    'Salary': [56000,25000,35000,28000,58000],
    'Bonus': [15000,7000,9000,13000,12000]
}, index=['E101','E102','E103','E104','E105'])

print(Employee)
```

```python
Employee.loc['E106'] = ['GEET', None, 'Analyst', 70000, 15000]
print(Employee)
```

```python
Employee['Shift'] = 'Morning'
print(Employee)
```

```python
print(Employee[Employee['Age'] == Employee['Age'].min()])
```

```python
print(Employee[Employee['Designation'] == 'Clerk'])
```
