```python
import pandas as pd

data = pd.DataFrame({
    'Count': [15, 20, 25, 35],
    'Price': [250, 200, 600, 900]
}, index=['Pencil', 'Pen', 'Eraser', 'Scale'])

print(data)
```

```python
data.loc['Eraser', 'Count'] = 15
print(data)
```

```python
data.rename(index={'Scale': 'Wooden Scale'}, inplace=True)
print(data)
```

```python
data.index = ['Product1', 'Product2', 'Product3', 'Product4']
print(data)
```

```python
data.columns = ['Discount', 'Total']
print(data)
```

```python
data.loc['Product5'] = [10, 100]
print(data)
```

```python
Products = data.drop(columns=['Total'])
print(Products)
```
